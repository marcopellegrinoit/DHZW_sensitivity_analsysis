```python
import subprocess
import os
import csv
import pandas as pd
from tqdm import tqdm
```

# Define the parameters and the values to try

```python
# Open the CSV file and read it as a list of dictionaries
with open('data/alpha_labels.csv', newline='') as csvfile:
    reader = csv.DictReader(csvfile)
    alphas = [row['parameter'] for row in reader]
with open('data/parameter_labels - full.csv', newline='') as csvfile:
    reader = csv.DictReader(csvfile)
    params = [row['parameter'] for row in reader]
```

```python
# Define the range of values for each parameter
values = [-4.0, -2.0, 0.0, 2.0, 4.0]
```

# Create the entier parameter sets

```python
# Create an empty table with the column headers from params
table = pd.DataFrame(columns=params)

table['used_parameter_label'] = 0
table['used_parameter_value'] = 0

table
```

```python
# create a table with all the parameters. only the alphas must vary
for alpha in alphas:
    for value in values:
        # Create a row with all zeros except for the current value of the loop
        zeros_row = pd.Series([value if p == alpha else 0.0 if p in alphas else 1.0 for p in params], index=params)
        zeros_row['used_parameter_label'] = alpha
        zeros_row['used_parameter_value'] = value
        
        table.loc[len(table)] = zeros_row
        
        #table = table.append(zeros_row, ignore_index=True)
```
